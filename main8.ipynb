```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Titanic-Dataset.csv')
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.shape
```

```python
x=df['Age'].mean()
df['Age'].fillna(x,inplace=True)
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
str(len(df))
```

```python
print('Visualization based on survived and non-survived ')
plt.figure()
target='Survived'
sns.countplot(data=df,x=target,hue='Survived').set_title("Survived and Non-Survived Passenger")
plt.show()
```

```python
print('Survived And Non Survived Visualization Based On Gender ')
plt.figure()
title='Survived'
sns.countplot(data=df,x=df['Sex'],hue=title).set_title('Survived And Non Survived Visualization Based On Gender ')
plt.show()
```

```python
print('Survived And Non-Survived Based On Class')
plt.figure()
sns.countplot(data=df,x='Survived',hue='Pclass').set_title('Survived And Non-Survived Based On Class')
plt.show()
```

```python
plt.figure()
sns.countplot(data=df,x='Survived',hue='Age').set_title('Survived and Non-survived based on age')
plt.show()
```

```python
plt.figure()
sns.countplot(data=df,x='Survived',hue='Fare')
plt.show()
```

```python
plt.figure()
sns.histplot(data=df,x='Age',bins=15,edgecolor='black')
plt.show()
```

```python
plt.figure()
sns.histplot(data=df,x='Fare',bins=30,edgecolor='black').set_title('Histogram Of Fare')
plt.show()
```
