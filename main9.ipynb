```python
''' Data Visualization II
1. Use the inbuilt dataset 'titanic' as used in the above problem.
Plot a box plot for distribution of age with respect to each
gender along with the information about whether they survived
or not. (Column names : 'sex' and 'age')
2. Write observations on the inference from the above
statistics. '''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Titanic-Dataset.csv')
```

```python
df.isnull().sum()
```

```python
x=df['Age'].mean()
df['Age'].fillna(x,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(15,15))
sns.boxplot(data=df,x='Sex',y='Age',hue='Survived').set_title('Representation of age wrt Gender ')
plt.legend(title='Survived',labels=['No','Yes'])
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.violinplot(data=df,x='Sex',y='Age',hue='Survived',split=True)
plt.title('Survived based on Gender and Age Group ')
plt.show()
```

```python
plt.figure(figsize=(20,20))
sns.swarmplot(data=df,x='Sex',y='Age',hue='Survived',dodge=True)
plt.title('Swarmplot Representation')
plt.show()
```
