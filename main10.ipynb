```python
cols=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
plt.figure(figsize=(10,4))
sns.histplot(data=df,x='SepalLengthCm',bins=20,edgecolor='black')
plt.title('Histogram Of SepalLengthCm')
plt.show()
```

```python
cols=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
plt.figure(figsize=(10,4))
sns.histplot(data=df,x='SepalWidthCm',bins=20,edgecolor='black')
plt.title('Histogram Of SepalWidthCm')
plt.show()
```

```python
cols=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
plt.figure(figsize=(10,4))
sns.histplot(data=df,x='PetalLengthCm',bins=20,edgecolor='black')
plt.title('Histogram Of PetalLengthCm')
plt.show()
```

```python
cols=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
plt.figure(figsize=(10,4))
sns.histplot(data=df,x='PetalWidthCm',bins=20,edgecolor='black')
plt.title('Histogram Of PetalWidthCm')
plt.show()
```

```python
plt.figure(figsize=(12,4))
sns.boxplot(data=df,x='SepalLengthCm',hue='Species')
plt.title('Boxplot Of Sepallength')
plt.show()
```

```python
plt.figure(figsize=(12,4))
sns.boxplot(data=df,x='SepalWidthCm',hue='Species')
plt.title('Boxplot Of SepalWidth')
plt.show()
```
